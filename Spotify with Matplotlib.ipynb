```python
#using Matplotlib to analyse spotify data from Kaggle
#Data Source: https://www.kaggle.com/leonardopena/top50spotify2019
```

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
spotifyData = pd.read_csv('/Users/kaykay/Downloads/Kaggle Data/top50.csv', index_col = 0, encoding='latin-1')
```

```python
spotifyData.head(5)
```

```python
#What makes a song popular?
spotifyData.info()
```

```python
x = spotifyData['Popularity']
y = spotifyData['Genre']
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_title('Favourite Genre')
axes.set_xlabel('Popularity')
axes.set_ylabel('Genre')
```

```python
#BarChart
fig1, ax1 = plt.subplots()
set_colors = ['#fcba03','#03f8fc','g', 'r', 'c', 'm', 'y', 'g']
ax1.barh(y, x, color = set_colors)

#Labels
ax1.set_title('Favourite Genre')
ax1.set_xlabel('Popularity')
ax1.set_ylabel('Genre')
```

```python
#Electropop is the favourite genre
```

```python
# Does loudness influence the best genre?

z = spotifyData['Loudness..dB..']

plt.barh(y, z)

```

```python
fig, axs = plt.subplots()
axs.scatter(z,x, alpha=0.5)
axs.set_title('Loudness vs Popularity')
axs.set_xlabel('Loudness')
axs.set_ylabel('Popularity')
```

```python
audioFeatures_df = spotifyData[['Energy', 'Danceability','Liveness','Valence.','Acousticness..', 'Speechiness.' ]] 
```

```python
audioFeatures_df.head(5)
```

```python
audioFeatures_df.mean().plot(color='r', marker="o")
plt.title('Audio Features Mean Values')
```

```python
# The plot shows danceabillity is very important and energy next
#Matplotlib histogram to see how danceability is distibuted
```

```python
plt.hist(audioFeatures_df['Danceability'], color = 'blue', edgecolor = 'black'
        )

plt.title('Histogram of Danceability')
plt.xlabel('Danceability')
```

```python
#It is clear danceability is right skewed


#Matplotlib histogram to see how Energy is distibuted
#Compared to musc is more evenely distributed
```

```python
plt.hist(audioFeatures_df['Energy'], color = 'yellow', edgecolor = 'black'
        )

plt.title('Histogram of Energy')
plt.xlabel('Energy')
```

```python
#It is clear people like energwtic songs more than calming songs. 
#They also prefer to stream songs they can dance to. 
```
