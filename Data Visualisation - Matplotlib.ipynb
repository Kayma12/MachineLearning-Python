```python
import matplotlib.pyplot as plt

```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
#11 data points from 0 - 5

y = x ** 2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL METHOD
plt.plot(x,y)

#xlabel
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#Multiplot on the same canvas:
# the number of rows | number of columns | plot you referring to
plt.subplot(1,2,1)
plt.plot(x,y,'g--')


plt.subplot(1,2,2)
plt.plot(y,x, 'p')
```

```python
#Object Orientated method : creating figure objects amd calling methods
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # left| bottom |width |height

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
#subplot we can specify the number of rows and columns
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
#FIGURE SIZE AN DPI
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')
ax.legend()

```

```python
#Customising how a plot looks
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF2C00', lw=1, ls='-', marker='o',markersize=20)
```
