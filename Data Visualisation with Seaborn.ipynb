```python
import seaborn as sns
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#Distribution plot - univariate set of observation (1 variable)
```

```python
sns.distplot(tips['total_bill'], bins=30)
```

```python
sns.distplot(tips['tip'])
```

```python
#jointplot allows you to combine two different distribution plot (bivariate two variables)
```

```python
sns.jointplot(x= 'tip',y='total_bill',data= tips, kind = 'kde')
```

```python
sns.jointplot(x= 'tip',y='total_bill',data= tips)
```

```python
#The scatter plot tells us more tips are given where majority of the bills are. 
#Thererfore because less people spent over 30 on the bill, there is also less tips above 8 pounds, 
#so this tells us tips are proportionate to the total_bill.
```

```python
#Pairplot essentially plot pairwise relationships across the dataframe for numerical data
```

```python
sns.pairplot(tips,hue='sex')
```

```python
#Kernel density estimation plots ---> KDE
```

```python
#Categorical Plots
```

```python
sns.barplot(x= 'sex',y= 'total_bill',data=tips, estimator = np.std )
```

```python
#The plot shows the total bill for males is higher than the total bill for females.
```

```python
sns.boxplot(x= 'day',y='total_bill',data=tips, hue='smoker')
```

```python
#People tend to have a higher bill on sunday than any other days. 
#Adding a hue we see non smokers have a higher bill on Friday
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue= 'sex', aspect=0.6, height=8)
```

```python
#This plot shows alot of information, even though Sunday may seem to have a higher 
#but this plot shows that in actually saturday has a higher spending
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True)
```

```python
#New dataset
flights=sns.load_dataset('flights')
flights.head()
```
